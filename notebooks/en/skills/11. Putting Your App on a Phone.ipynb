# **Putting Your App on a Phone**

In the Gradio lesson, you built web apps and shared them with a link. But that link has a problem: it stops working when you close your Colab notebook, and even at best it only lasts a few days.

In this lesson, you'll give your app a permanent home on the internet, for free, and then install it on a phone so it works like a regular app, with its own icon on the home screen. By the end, you'll be able to hand someone your spellchecker as an app they can keep.

You won't need to learn any new programming. Everything here is clicking through websites, plus code you already wrote in earlier lessons.

## **What we're building toward**

1. Your app runs permanently on **Hugging Face Spaces**, a free hosting service made for Gradio apps.
2. Anyone can open it on any phone or computer with a normal web link.
3. With one extra setting, phones treat it as an installable app: it gets an icon, opens full screen, and lives on the home screen like anything from an app store.

This kind of installable website is called a **Progressive Web App** (PWA). It's how many real products ship their apps, and it means you don't need an Apple or Google developer account, app store approval, or any extra code.

## **Step 1: Test your app in Colab first**

We'll use a small spellchecker app as our example. It downloads the Uspanteko corpus, builds a lexicon, and flags words that aren't in it. Run it here first to make sure it works. You can replace it with any app you built in the earlier lessons.

```python
%pip install -q gradio
```

```python
# Download the helper code for this course from GitHub
import urllib.request

urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/lecs-lab/belt/main/notebooks/en/util.py",
    "util.py")

import util
import gradio as gr

# Build a lexicon from the sample corpus, like in the Spellchecker project
corpus_directory = util.download_corpus("usp")
corpus = util.load_raw_text(corpus_directory)
lexicon = set(util.tokenize(util.strip_accents(corpus).lower()))

def check_spelling(text):
    words = util.tokenize(util.strip_accents(text).lower())
    unknown = sorted(set(word for word in words if word not in lexicon))
    if not unknown:
        return "All words found in the lexicon!"
    return "Not in the lexicon: " + ", ".join(unknown)

interface = gr.Interface(
    fn=check_spelling,
    inputs=gr.Textbox(label="Type some Uspanteko text"),
    outputs=gr.Textbox(label="Result"),
    title="Uspanteko Spellchecker")

interface.launch(share=True)
```

## **Step 2: Create a Hugging Face account and a Space**

[Hugging Face](https://huggingface.co) is a company that hosts machine learning apps and models. Their free hosting for apps is called **Spaces**, and it supports Gradio directly.

1. Go to [huggingface.co/join](https://huggingface.co/join) and create a free account.
2. Click your profile picture in the top right and choose **New Space**.
3. Give your Space a name, for example `uspanteko-spellchecker`. This becomes part of your app's web address.
4. Under **Select the Space SDK**, choose **Gradio**.
5. Leave the hardware on the free option (**CPU basic**) and keep the Space **Public**, then click **Create Space**.

## **Step 3: Add your code to the Space**

A Space runs whatever is in a file called `app.py`. You can edit it right in the browser.

1. In your new Space, open the **Files** tab.
2. Click **Contribute** -> **Create a new file** (or edit `app.py` if one exists).
3. Name the file `app.py`, and paste in the code below. It's the same app as Step 1, with two small changes: the launch line turns on the app setting (`pwa=True`), and we don't need `share=True` because the Space already has a public address.
4. Click **Commit new file to main** at the bottom.

The Space will show **Building** for a minute or two, then your app appears, running at a permanent address.

```python
# This cell is for copying, not running: paste it into app.py on your Space

import urllib.request

urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/lecs-lab/belt/main/notebooks/en/util.py",
    "util.py")

import util
import gradio as gr

corpus_directory = util.download_corpus("usp")
corpus = util.load_raw_text(corpus_directory)
lexicon = set(util.tokenize(util.strip_accents(corpus).lower()))

def check_spelling(text):
    words = util.tokenize(util.strip_accents(text).lower())
    unknown = sorted(set(word for word in words if word not in lexicon))
    if not unknown:
        return "All words found in the lexicon!"
    return "Not in the lexicon: " + ", ".join(unknown)

interface = gr.Interface(
    fn=check_spelling,
    inputs=gr.Textbox(label="Type some Uspanteko text"),
    outputs=gr.Textbox(label="Result"),
    title="Uspanteko Spellchecker")

interface.launch(pwa=True)
```

## **Step 4: Install it on a phone**

Your app now has a permanent address of its own, separate from the Space's project page. To find it, click the **three dots** menu at the top right of your Space and choose **Embed this Space**. The **Direct URL** shown there (it ends in `.hf.space`) is your app's address. That's the link to share.

On the phone:

1. Open the direct URL in the phone's browser.
2. **iPhone (Safari):** tap the Share button, then **Add to Home Screen**.
3. **Android (Chrome):** tap the three dots menu, then **Add to home screen** (on some phones it says **Install app**).

The app now has its own icon and opens full screen, like any installed app.

Two honest limitations to know about:

- **It needs internet.** The app runs on the Space, not on the phone itself.
- **Free Spaces go to sleep** after about two days without visitors. The app still works, but the first person to open it after a quiet period waits a minute while it wakes up.

## **What about the app stores?**

Publishing in the Apple App Store or Google Play Store is a much bigger undertaking: developer accounts (Apple's costs 99 USD per year), app review processes, and new programming tools. For sharing a language tool with your community, the installable web app you just made does almost everything a store app would, with none of that overhead.

If your community needs an app that works **without internet**, that's a genuinely harder project (the language model has to run on the phone itself), and a good thing to bring to the course forum or a collaborating developer.

## **Summary**

- Colab share links are temporary; **Hugging Face Spaces** gives your Gradio app a free permanent home
- A Space is just your same app code in a file called `app.py`, edited right in the browser
- `launch(pwa=True)` makes the app installable: **Add to Home Screen** puts it on a phone with its own icon
- The app needs internet, and free Spaces nap when nobody visits, but for sharing tools with your community this is the fastest real path from notebook to phone
