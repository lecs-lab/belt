```python
import os
import re
corpus = ""

for file_name in os.listdir("../../corpora/esp/evltest"):
    # Read the file as a string
    file_path = os.path.join("../../corpora/esp/evltest", file_name)
    if os.path.isdir(file_path):
        continue
        
    #  Make sure we only read text files
    if ".txt" not in file_name:
        continue
        
    with open(file_path, 'r', encoding="ISO-8859-1") as file:
        with open(file_path + "fixed.txt", "w") as out:
            for line in file.readlines():
                line = line[17:]
                if line[:2] == "//":
                    continue
                line = re.sub(r"\[\w*\]", "", line)
                line = re.sub(r"\{\w*\}", "", line)
                line = re.sub(r"\<[\w\s]*\>", "", line)
                out.write(line)
```
